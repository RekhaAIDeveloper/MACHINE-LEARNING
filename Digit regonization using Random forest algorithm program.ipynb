Importing Basic Libraries

```python
import pandas as pd
```

Load Dataset

```python
data=pd.read_csv('train.csv.xls')
data
```

Summarize Dataset

```python
print(data.shape)
```

Mapping Text Data to Binary Value

Segregate Dataset into X(Input/Independent Variable) and Y(Output/Dependent Variable)

```python
X=data.iloc[:,1:]
X
```

```python
Y=data.iloc[:,0]
Y
```

Test Again to Check any na value

Splitiing Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=2)
```

Display the train and test data

```python
x_train
```

```python
y_train
```

#  Training

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

Prediction for all Test Data

```python
y_pred=model.predict(x_test)
```

Accuracy of Our Model

```python
from sklearn.metrics import accuracy_score
print('Accuracy of my Model:',accuracy_score(y_test,y_pred)*100,'%')
```
